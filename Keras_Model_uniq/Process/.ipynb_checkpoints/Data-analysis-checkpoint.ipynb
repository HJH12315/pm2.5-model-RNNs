```python
from __future__ import print_function, division
import pandas as pd
import numpy as np
from numpy import log

import os
import math
from pandas import Series, DataFrame, Panel
import re
import random
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler
from sklearn.decomposition import PCA
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose

np.set_printoptions(threshold=np.nan)
np.set_printoptions(precision=3)
np.set_printoptions(suppress=True)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def to_timeseries(df,col_name,col_num,fromtime,totime,freq):
    dates = pd.date_range(fromtime, totime, freq=freq)
    df_np = df.as_matrix()
    df_ts = Series(df_np[:,col_num], index=dates)
    df_ts = DataFrame({col_name : df_ts})
    df_ts = df_ts.astype(float)
    return df_ts
```

```python
data_loaded = pd.read_pickle("mumbai_6.pkl")

data_loaded_pm_ts = to_timeseries(data_loaded,'PM2.5',3,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_ws_ts = to_timeseries(data_loaded,'WS',4,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_rh_ts = to_timeseries(data_loaded,'RH',5,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_bp_ts = to_timeseries(data_loaded,'BP',6,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_vws_ts = to_timeseries(data_loaded,'VWS',7,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_sr_ts = to_timeseries(data_loaded,'SR',8,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_wd_ts = to_timeseries(data_loaded,'WD',9,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')
data_loaded_temp_ts = to_timeseries(data_loaded,'TEMP',10,'2015-01-01 00:00:00','2017-03-31 23:00:00','H')

data_loaded_pm_ts.plot()
data_loaded_pm_ts.hist()

data_loaded_ws_ts.plot()
data_loaded_ws_ts.hist()

data_loaded_rh_ts.plot()
data_loaded_rh_ts.hist()

data_loaded_bp_ts.plot()
data_loaded_bp_ts.hist()

data_loaded_vws_ts.plot()
data_loaded_vws_ts.hist()

data_loaded_sr_ts.plot()
data_loaded_sr_ts.hist()

data_loaded_wd_ts.plot()
data_loaded_wd_ts.hist()

data_loaded_temp_ts.plot()
data_loaded_temp_ts.hist()
```
